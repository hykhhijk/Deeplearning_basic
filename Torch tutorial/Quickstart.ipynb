# import Library&Datasets

```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
```

```python
training_data = datasets.FashionMNIST(
root="data", train=True, download=True, transform=ToTensor())

test_data = datasets.FashionMNIST(
root="data", train=False, download=True, transform=ToTensor())
```

## iterate dataset 

```python
batch_size = 64

train_dataloader = DataLoader(training_data, batch_size = batch_size)
test_dataloader = DataLoader(test_data, batch_size = batch_size)
                             
for X, y in test_dataloader:
    print("Shape of X [N, C, H, W] {}".format(X.shape))
    print("Shaper of y: {}".format(y.shape))
    break
```

# Making model 

```python
device = "cuda"if torch.cuda.is_available() else "cpu"
print("Using device: {}".format(device))

### __init__에서 연산 정의
### __forward를 통해 데이터 전달법 정의?


class NN(nn.Module):
    def __init__(self):
        super(NN, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
        nn.Linear(28*28, 512),
        nn.ReLU(),
        nn.Linear(512, 512),
        nn.ReLU(),
        nn.Linear(512, 10)
        )
        
    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits

model = NN().to(device)
print(model)
```

## Fitting

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
```

```python
for batch, (X, y) in enumerate(train_dataloader):
    print(batch, X.shape, y.shape)
# means batch: batch index, X: batch of data, y: label of data batch
```

```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    #size = 60000
    
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)
        
        pred=model(X)
        loss = loss_fn(pred, y)
        
        ##########################
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        ########################### -->?? What this mean?
        
        
        if batch % 100 ==0:
            loss, current = loss.item(), batch * len(X)
            print("loss: {} [{} / {}]".format(loss, current, size))
```

```python
def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1)==y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print("Test error: \n Accuraqcy: {}%, Avg loss: {}\n".format(100*correct, test_loss))
```

```python
epochs = 5
for t in range(epochs):
    print("Epoch {}\n------------------------------------".format(t+1))
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)
print("Done!!")
```

## Save model 

```python
torch.save(model.state_dict(), "model.pth")
print("Saved Pytorch Model State to model.pth")
```
